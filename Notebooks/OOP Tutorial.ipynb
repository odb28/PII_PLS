```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0,30,1)
a = lambda x: 30 - 30/np.exp(x/10)
y = a(x)
print(y)
q = [30]*30
```

```python
fig, ax0 = plt.subplots()
ax0.plot(x,y, label = "Observed count")
ax0.plot(x,q,linestyle = "--", label = "Estimated asymptote")
ax0.set_xlabel("Sampling effort")
ax0.set_ylabel("Species Richness")
ax0.set_title("Figure 1")
ax0.legend()
fig.show()
```
