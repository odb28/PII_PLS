```python
import numpy as np
from src.baseSIR import model_sir
from src.baseSIR import cull_sir
import scipy as sp
import arviz as az
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
import time
```

```python
N_10_b = np.genfromtxt(f"../../Data/fixed_G_N_10_b.csv", delimiter= ",")
N_100_b = np.genfromtxt(f"../../Data/fixed_G_N_100_b.csv", delimiter= ",")
N_1000_b = np.genfromtxt(f"../../Data/fixed_G_N_1000_b.csv", delimiter= ",")
gamma = 1
```

```python
seed = 1912
rng = np.random.default_rng(seed)

mu = 0
tmax = 10
tstep = 0.05
n_sim = 1000

final_sizes = pd.DataFrame(np.empty(shape=(30,2)),columns=("S","N"))
size_stds = pd.DataFrame(np.empty(shape=(30,2)),columns=("S_std","N"))
pr_ext= pd.DataFrame(np.empty(shape=(30,2)),columns=("Pr_ext","N"))
for i in range(30):
    final_s = []
    exts = []

    if i < 10:
        final_sizes.iloc[i,1] = 10
        size_stds.iloc[i,1] = 10
        pr_ext.iloc[i,1] = 10
        X0 = [9, 1, 0]
        for j in range(n_sim):
            s,ext = model_sir(X0,mu,beta=N_10_b[i,0],gamma=gamma,tmax=tmax,tstep=tstep,rng=rng)
            if ext == 0:
                final_s.append(s*100)
            exts.append(ext)

    elif 10 <= i < 20:
        final_sizes.iloc[i,1] = 100
        size_stds.iloc[i,1] = 100
        pr_ext.iloc[i,1] = 100
        X0 = [90, 10, 0]
        for j in range(n_sim):
            s,ext = model_sir(X0,mu,beta=N_100_b[i-10,0],gamma=gamma,tmax=tmax,tstep=tstep,rng=rng)
            if ext == 0:
                final_s.append(s*10)
            exts.append(ext)

    elif i >= 20:
        final_sizes.iloc[i,1] = 1000
        size_stds.iloc[i,1] = 1000
        pr_ext.iloc[i,1] = 1000
        X0 = [900, 100, 0]
        for j in range(n_sim):
            s,ext = model_sir(X0,mu,beta=N_1000_b[i-20,0],gamma=gamma,tmax=tmax,tstep=tstep,rng=rng)
            if ext == 0:
                final_s.append(s)
            exts.append(ext)

    final_sizes.iloc[i,0] = np.mean(final_s)
    pr_ext.iloc[i,0] = sum(exts)/n_sim
    size_stds.iloc[i,0] = np.std(final_s)
np.savetxt("../../Data/Final_size_no_c_fixed_G.csv",final_sizes,delimiter=",")
np.savetxt("../../Data/Pr_ext_no_c_fixed_G.csv",pr_ext,delimiter=",")
```
