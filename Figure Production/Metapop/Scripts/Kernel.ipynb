```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
kernel = lambda x: 1/x**2
```

```python
distances = np.arange(1,20,1)
rels = kernel(distances)
long_distances = np.arange(1,1000,1)
long_rels = kernel(long_distances)
```

```python
fig, ax0 = plt.subplots()
ax0.plot(distances,rels,color="maroon")
ax0.scatter(10,kernel(10),color="b")
ax0.scatter(np.sqrt(200),kernel(np.sqrt(200)),color="b")
ax0.annotate(text="Nodes 2&3", xy= (10,kernel(10)), xytext= (10,0.2), arrowprops=dict(facecolor='black', shrink=0.05))
ax0.annotate(text="Node 4", xy= (np.sqrt(200),kernel(np.sqrt(200))), xytext= (np.sqrt(200),0.2), arrowprops=dict(facecolor='black', shrink=0.05))
ax0.set_ylabel("Force of Infection")
ax0.set_xlabel("Distance between Nodes")
ax0.set_title("Dispersal Kernel")
ins = ax0.inset_axes([0.5,0.5,0.5,0.5])
ins.plot(long_distances,np.log(long_rels),color="maroon")
ins.set_ylabel("Log(Force of Infection)")
ins.set_xlabel("Distance between Nodes")
fig.show()
```
